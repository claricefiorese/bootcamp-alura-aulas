### Desafio 03: Classificar algumas perguntas como Qualitativa x Quantitativa 

Variáveis quantitativas: são as características que podem ser medidas em uma escala quantitativa (de medidas), ou seja, apresentam valores numéricos ou resultantes de contagem. Podem ser contínuas ou discretas.
- Discretas: são características mensuráveis que podem assumir apenas um número finito ou enumerável de valores e, assim, somente fazem sentido valores inteiros. Geralmente são o resultado de contagens.
- Contínuas: são características mensuráveis que assumem valores em um intervalo, aberto ou fechado, de números reais, para as quais valores fracionais fazem sentido. Usualmente devem ser medidas através de algum instrumento.

Variáveis qualitativas ou categóricas: são as características que não possuem valores quantitativos, não podem ser mensuradas numericamente, mas, ao contrário, são definidas por categorias, ou seja, representam uma classificação, qualidade ou descrição dos indivíduos. Podem ser nominais ou ordinais.
- Nominais: quando as variáveis não podem ser hierarquizadas ou ordenadas, sendo comparadas apenas por igualdade ou diferença.
- Ordinais: quando as variáveis mantém uma relação relação de ordem, mas não é possível quantificar a diferença entre os resultados.
